```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
import pandas as pd
import copy

import belief_propagation as bp
```

```python
colv = np.loadtxt('COLVAR')
colv = colv[:,1:] # removing the time column
```

```python
colv_df = pd.DataFrame(colv, columns=('$\phi_1$', '$\psi_1$', '$\phi_2$', '$\psi_2$', '$\phi_3$', '$\psi_3$'))
colv_df
```

## Here I show the factor graph as we vary the graphical lasso penalty

```python
alphas = np.logspace(-1,0,10)
for a in alphas:
    print('ALPHA = '+ str(a))
    FG = bp.FG_from_glasso(colv_df,a)
    funcs = [x for x,y in FG.nodes(data=True) if y['bipartite']==1]
    print('DEGREES')
    print([(x,FG.degree[x]) for x,y in FG.nodes(data=True) if y['bipartite'] ==1 ])
    max_degree = np.max([FG.degree[f] for f in funcs])
    print('MAX FACTOR DEGREE = ' + str(max_degree))
    bp.show_FG_components(FG,figsize=(8,8))
    plt.show()
    print('====================================')
```

## This is the factor graph we actually use

```python
a = np.log(1.14632)
figsize=(8,8)
print('ALPHA = '+ str(a))
FG = bp.FG_from_glasso(colv_df,a)
funcs = [x for x,y in FG.nodes(data=True) if y['bipartite']==1]
print('DEGREES')
print([(x,FG.degree[x]) for x,y in FG.nodes(data=True) if y['bipartite'] ==1 ])
max_degree = np.max([FG.degree[f] for f in funcs])
print('MAX FACTOR DEGREE = ' + str(max_degree))
plt.show()
print('====================================')
 
    
    
components = [FG.subgraph(c).copy() for c in nx.connected_components(FG)]

for C in components:
    plt.figure(figsize=figsize)
    pos = nx.kamada_kawai_layout(C)
    var = [x for x,y in C.nodes(data=True) if y['bipartite']==0]
    func = [x for x,y in C.nodes(data=True) if y['bipartite']==1]
    nx.draw_networkx_nodes(C, nodelist = var, pos=pos, node_size=1500)
    nx.draw_networkx_nodes(C, nodelist = func, pos=pos, node_shape='s', node_size=1500)
    nx.draw_networkx(C,pos=pos,with_labels=True,node_size=0, font_color='w',font_size=18)
```

## Message passing in a random order
## We quickly see the cross entropy fall in the first pass over each variable

```python
MP =  bp.MessagePasserPeriodic(FG, colv_df[::50])
```

```python
f_list = copy.copy(MP.funcs)
for i in range(3):
    np.random.shuffle(f_list)

    for f in f_list:
        bp.update_show_if_2d(MP, f)
        print('CE: ' + str(MP.global_CE()))
          
    print('='*110)
```
